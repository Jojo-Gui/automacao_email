```python
# Importando Biblioteca
import pandas as pd
import numpy as np
import pyarrow.parquet as pq

# Configuração Global Pandas - formata float com 2 casas decimais
pd.set_option('display.float_format', '{:.2f}'.format)

# Configuração Global Pandas - mostra todas as colunas do df
pd.set_option('display.max_columns', None)

```

```python
# Carregando base de dados
dfm = pd.read_parquet("../data/amazon_tratada.parquet")
dfm
```

```python
dfm.info()
```

```python
# Quantidade de produtos únicos
prod_uniq = dfm['product_name'].nunique()
prod_uniq
```

```python
# Contando quantidade de produtos por categorias principais

qtd_produtos = (dfm
                .groupby('Category n0')['product_id']
                .nunique()
                .reset_index(name='qtd_produtos')
                .sort_values(by='qtd_produtos', ascending=False).iloc[0]
                )

print(qtd_produtos['qtd_produtos'])
```

```python
# Considerações 1

print(f'Há ao todo {prod_uniq} produtos únicos na base de dados, distribuídos em {dfm["Category n0"].nunique()} categorias principais, onde a categoria com mais produtos é a \n{qtd_produtos["Category n0"]} com {qtd_produtos["qtd_produtos"]} produtos.')
```

```python
# Verificando preço máximo e mínimo
dfm.describe()['actual_price'].iloc[[3,7]]
```

```python
# Produtos mais caro e mais barato antes do desconto 

prod_mais_caro = dfm.loc[dfm['actual_price'].idxmax(), 'product_name']
prod_mais_barato = dfm.loc[dfm['actual_price'].idxmin(), 'product_name']
preço_prod_mais_caro = dfm['actual_price'].max()
preço_prod_mais_barato = dfm['actual_price'].min()

print(prod_mais_caro)
print(preço_prod_mais_caro)
print(prod_mais_barato)
print(preço_prod_mais_barato)
```

```python
dfm.describe()['discounted_price'].iloc[1]
```

```python
# Comparando a média e a mediana dos produtos com desconto

dfm = dfm.sort_values(by='discounted_price', ascending=False)
media_desc = dfm['discounted_price'].mean()
mediana_desc = dfm['discounted_price'].median()

print(media_desc)
print(mediana_desc)
```

```python
# Considerações 2

print(f'O produto mais caro antes do desconto é o {prod_mais_caro} com o valor \nde {preço_prod_mais_caro:.2f}, enquanto o produto mais barato é o {prod_mais_barato} com o valor de {preço_prod_mais_barato:.2f}. \nA média dos preços com desconto é de {media_desc:.2f} e a mediana é de {mediana_desc:.2f}, indicando que os preços estão relativamente distantes, \nmas a média é um pouco maior devido a presença de alguns produtos com descontos mais altos, o que é comum em e-commerces a \nmédia ser bem distante da mediana pela variedade de produtos.')
```

```python
dfm.columns
```

```python
# Produto com maior desconto percentual

prod_maior_desc_p = dfm.loc[dfm['discount_percentage'].idxmax(), 'product_name']
valor_maior_desc_p = dfm[['actual_price', 'discount_percentage','discounted_price', 'full_discount']][dfm['product_name'] == prod_maior_desc_p]

valor_maior_desc_p
```

```python
# Produto com maior desconto valor (full_discount)

prod_maior_desc_v = dfm.loc[dfm['full_discount'].idxmax(), 'product_name']
valor_maior_desc_v = dfm[['actual_price', 'discount_percentage','discounted_price', 'full_discount']][dfm['product_name'] == prod_maior_desc_v]

valor_maior_desc_v
```

```python
# Comparativo valor_maior_desc_v com valor_maior_desc_p

comparativo = pd.concat([valor_maior_desc_v, valor_maior_desc_p])
comparativo
```

```python
# Média de descontos por categoria

med_desc_cat = (dfm.groupby('Category n0')['discount_percentage'].mean().reset_index(name='media_desconto').sort_values(by='media_desconto', ascending=False))

med_desc_cat
```

```python
# Considerações 3

print(f'O produto com maior desconto percentual é o \n{prod_maior_desc_p} \ncom um desconto de \n{valor_maior_desc_p["discount_percentage"].values[0]*100:.2f}% e uma economia real de {valor_maior_desc_p["full_discount"].values[0]:.2f}, enquanto o produto com maior desconto em valor é o \n{prod_maior_desc_v} com um desconto de {valor_maior_desc_v["discount_percentage"].values[0]*100:.2f}% e uma economia real de {valor_maior_desc_v["full_discount"].values[0]:.2f}.')
print(f'A categoria com o maior desconto médio é a {med_desc_cat.iloc[0]['Category n0']} com um desconto médio de {med_desc_cat.iloc[0]['media_desconto']*100:.2f}%.')
```

```python
# Média de avaliações por categoria

cat_avaliation_mean = dfm.groupby('Category n0')['rating'].mean().reset_index(name='media_avaliaçãoes').sort_values(by='media_avaliaçãoes', ascending=False)
cat_avaliation_mean
```

```python
# Top 5 produtos com mais avaliazaçãoes

top_5_rating_count = dfm[['product_name','Category n0','rating_count']].sort_values(by='rating_count', ascending=False).head(5)
top_5_rating_count
```

```python
# Considerações 4 - Qual categoria tem a melhor avaliação média (rating)? Quais os 5 produtos com mais avaliações (rating_count)?

print(f'A categoria com o maior desconto médio é a {med_desc_cat.iloc[0]["Category n0"]} com um desconto médio de {med_desc_cat.iloc[0]["media_desconto"]*100:.2f}%.')
print(f'Os cinco produtos com maior avaliação são:\n {top_5_rating_count[['product_name','rating_count']]}')
```
